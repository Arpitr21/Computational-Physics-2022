```python
import matplotlib.pyplot as plt
import numpy as np
theta=np.linspace(0,np.pi,100)
x = [(10 + 10*(2*(np.tan(i)**2)+ 1)**0.5)*(np.cos(i))**2 for i in theta]
plt.plot(theta,x)
plt.xlabel('theta')
plt.ylabel('x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x, y):
    return (2*np.sin((x)/2)*np.sin((y)/3))/(6*np.pi*np.pi)**0.5
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title('n_x = 1, n_y = 1')
surface_plot = ax.plot_surface(X, Y, Z)
def f(x, y):
    return (2*np.sin((x)/2)*np.sin((2*y)/3))/(6*np.pi*np.pi)**0.5
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title('n_x = 1, n_y = 2')
surface_plot = ax.plot_surface(X, Y, Z)

def f(x, y):
    return (2*np.sin((2*x)/2)*np.sin((y)/3))/(6*np.pi*np.pi)**0.5
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title('n_x = 2, n_y = 1')
surface_plot = ax.plot_surface(X, Y, Z)

def f(x, y):
    return (2*np.sin((2*x)/2)*np.sin((2*y)/3))/(6*np.pi*np.pi)**0.5
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title('n_x = 2, n_y = 2')
surface_plot = ax.plot_surface(X, Y, Z)
plt.show()
```
